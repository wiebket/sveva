# Analysis Switchboard Cellular Part 1 & 2


## Preprocessing


```python
# preprocessing metadata

print("Switchboard Cellular 1 Info")
print("Switchboard Cellular 1 Spkrinfo Length", len(switchboardCellular1_spkrinfo_df))
print("Switchboard Cellular 1 Callinfo Length", len(switchboardCellular1_callinfo_df))
print("Switchboard Cellular 1 Audit Length", len(switchboardCellular1_audit_df))
print("Switchboard Cellular 1 Phoneinfo Length", len(switchboardCellular1_phone_df))

switchboardCellular1_df = pd.merge(switchboardCellular1_spkrinfo_df, switchboardCellular1_audit_df, on="Pin")
switchboardCellular1_df = pd.merge(switchboardCellular1_df, switchboardCellular1_callinfo_df, on="CallID")

switchboardCellular1_callee_phone_df = switchboardCellular1_phone_df.rename(columns={"PhoneID": "CalleePhoneID",
                                                                                     "ServiceProvider": "CalleeServiceProvider",
                                                                                     "ServiceType": "CalleeServiceType",
                                                                                     "PhoneManufacturer": "CalleePhoneManufacturer",
                                                                                     "PhoneModel": "CalleePhoneModel",
                                                                                     "Phone": "CalleePhone"})
switchboardCellular1_caller_phone_df = switchboardCellular1_phone_df.rename(columns={"PhoneID": "CallerPhoneID",
                                                                                     "ServiceProvider": "CallerServiceProvider",
                                                                                     "ServiceType": "CallerServiceType",
                                                                                     "PhoneManufacturer": "CallerPhoneManufacturer",
                                                                                     "PhoneModel": "CallerPhoneModel",
                                                                                     "Phone": "CallerPhone"})

# Phone ID in callee_phone can be A or B with different mobile phone models, for simplicity keep only one of those
switchboardCellular1_callee_phone_df["CalleePhoneID"] = switchboardCellular1_callee_phone_df["CalleePhoneID"].replace({"0133A": "0133", "0133B": "0133", "0138A": "0138", "0138B": "0138", "0203A": "0203", "0203B": "0203", "0390A": "0390", "0390B": "0390", "0425A": "0425", "0425B": "0425", "0441A": "0441", "0441B": "0441", "0458A": "0458", "0458B": "0458", "0514A": "0514", "0514B": "0514", "0516A": "0516", "0516B": "0516", "0541A": "0541", "0541B": "0541"})

switchboardCellular1_callee_phone_df = switchboardCellular1_callee_phone_df.drop_duplicates(subset=["CalleePhoneID"])
switchboardCellular1_callee_phone_df["CalleePhoneID"] = switchboardCellular1_callee_phone_df["CalleePhoneID"].astype(int)

switchboardCellular1_caller_phone_df["CallerPhoneID"] = switchboardCellular1_caller_phone_df["CallerPhoneID"].replace({"0133A": "0133", "0133B": "0133", "0138A": "0138", "0138B": "0138", "0203A": "0203", "0203B": "0203", "0390A": "0390", "0390B": "0390", "0425A": "0425", "0425B": "0425", "0441A": "0441", "0441B": "0441", "0458A": "0458", "0458B": "0458", "0514A": "0514", "0514B": "0514", "0516A": "0516", "0516B": "0516", "0541A": "0541", "0541B": "0541"})

switchboardCellular1_caller_phone_df = switchboardCellular1_caller_phone_df.drop_duplicates(subset=["CallerPhoneID"])
switchboardCellular1_caller_phone_df["CallerPhoneID"] = switchboardCellular1_caller_phone_df["CallerPhoneID"].astype(int)

switchboardCellular1_df["CalleePhoneID"] = switchboardCellular1_df["CalleePhoneID"].replace({"0516?": "0516", "0458?": "0458", "0390?": "0390", "0203?": "0203", "0441?": "0441", "0541?": "0541", "0133?": "0133", "0425?": "0425", "0138?": "0138"})
switchboardCellular1_df["CallerPhoneID"] = switchboardCellular1_df["CallerPhoneID"].replace({"0516?": "0516", "0514?": "0514", "0458?": "0458", "0390?": "0390", "0133?": "0133", "0203?": "0203"})

switchboardCellular1_df["CalleePhoneID"] = switchboardCellular1_df["CalleePhoneID"].astype(int)
switchboardCellular1_df["CallerPhoneID"] = switchboardCellular1_df["CallerPhoneID"].astype(int)

switchboardCellular1_df = pd.merge(switchboardCellular1_df, switchboardCellular1_callee_phone_df, on="CalleePhoneID")
switchboardCellular1_df = pd.merge(switchboardCellular1_df, switchboardCellular1_caller_phone_df, on="CallerPhoneID")

print("Switchboard Cellular 1 Length", len(switchboardCellular1_df))

print("Switchboard Cellular 2 Info")
print("Switchboard Cellular 2 Spkrinfo Length", len(switchboardCellular2_spkrinfo_df))
print("Switchboard Cellular 2 Callinfo Length", len(switchboardCellular2_callinfo_df))
print("Switchboard Cellular 2 Audit Length", len(switchboardCellular2_audit_df))
print("Switchboard Cellular 2 Phoneinfo Length", len(switchboardCellular2_phone_df))

# PIN Number 5202 is an duplicate in Speaker-Info, Dropping one so that the number of calls matches 4040
switchboardCellular2_df = pd.merge(switchboardCellular2_spkrinfo_df.drop_duplicates(subset="Pin"), switchboardCellular2_audit_df, on="Pin")
switchboardCellular2_df = pd.merge(switchboardCellular2_df, switchboardCellular2_callinfo_df, on="CallID")

switchboardCellular2_callee_phone_df = switchboardCellular2_phone_df.rename(columns={"PhoneID": "CalleePhoneID",
                                                                                     "ServiceProvider": "CalleeServiceProvider",
                                                                                     "ServiceType": "CalleeServiceType",
                                                                                     "PhoneManufacturer": "CalleePhoneManufacturer",
                                                                                     "PhoneModel": "CalleePhoneModel",
                                                                                     "Phone": "CalleePhone"})
switchboardCellular2_caller_phone_df = switchboardCellular2_phone_df.rename(columns={"PhoneID": "CallerPhoneID",
                                                                                     "ServiceProvider": "CallerServiceProvider",
                                                                                     "ServiceType": "CallerServiceType",
                                                                                     "PhoneManufacturer": "CallerPhoneManufacturer",
                                                                                     "PhoneModel": "CallerPhoneModel",
                                                                                     "Phone": "CallerPhone"})

switchboardCellular2_df["CalleePhoneID"] = switchboardCellular2_df["CalleePhoneID"].replace({"0458?": "0458", "0390?": "0390", "0203?": "0203", "0516?": "0516"})

switchboardCellular2_df["CalleePhoneID"]= switchboardCellular2_df["CalleePhoneID"].astype(int)

switchboardCellular2_df = pd.merge(switchboardCellular2_df, switchboardCellular2_callee_phone_df, on="CalleePhoneID")
switchboardCellular2_df = pd.merge(switchboardCellular2_df, switchboardCellular2_caller_phone_df, on="CallerPhoneID")

print("Switchboard Cellular 2 Length", len(switchboardCellular2_df))

# Combine Switchboard Cellular datasets

switchboardCellular1_df["Set"] = "Cellular 1"
switchboardCellular2_df["Set"] = "Cellular 2"
switchboardCellular_df = pd.concat([switchboardCellular1_df, switchboardCellular2_df])

# rename "unk" to 0
switchboardCellular_df.loc[switchboardCellular_df["Age"] == "unk", "Age"] = 0
switchboardCellular_df["Age"]= switchboardCellular_df["Age"].astype(np.int64)
switchboardCellular_df.loc[switchboardCellular_df["Age"] == 0, "Age"] = np.nan

switchboardCellular1_df.head()
```
